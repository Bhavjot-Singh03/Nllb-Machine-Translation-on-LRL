```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', None)
```

```python
HEKP_dataset= pd.read_csv("/kaggle/input/parallel-hindi-english-data/Kinnauri_dataset.csv")
HEK_dataset= pd.read_csv("/kaggle/input/parallel-hindi-english-data/Kangri_dataset.csv")

KP_dataset= pd.read_csv("/kaggle/input/isolated-dataset/Kinnauri.csv")
K_dataset= pd.read_csv("/kaggle/input/isolated-dataset/Kangri.csv")
```

```python
Hin_Eng_Kin = pd.concat([HEKP_dataset, KP_dataset], axis=1)
Hin_Eng_Kan = pd.concat([HEK_dataset, K_dataset], axis=1)
```

```python
Hin_Eng_Kin.to_csv('/kaggle/working/Kinnauri_final.csv', index=False)
Hin_Eng_Kan.to_csv('/kaggle/working/Kangri_final.csv', index=False)
```
